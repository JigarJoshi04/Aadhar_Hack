- Age => 18 to 30 (15% of this data)
- Age => 30 to 50 (75% of this data)
- Age => 50 to 80 (10% of this data)


Division of sales data as per quality
Premium:
- 50% of total data will be for premium product
Mid- Range:
- 50% of total data will be for medium range product

Division of sales data as per income
Premium:
- Monthly Income range => 100k to 200k (80% of this data)
- Monthly Income range => 80k to 100k (10% of this data)
- Monthly Income range => 60k to 80k (5% of this data)
- Monthly Income range => 40k to 60k (5% of this data)
Mid-Range:
- Monthly Income range => 40k to 60k (80% of this data)
- Monthly Income range => 20k to 40k (15% of this data)
- Monthly Income range => 10k to 20k (5% of this data)

Division of sales data as per gender
Premium:
- Male => 5% of this data
- Female => 95% of this data
Mid-Range:
- Male => 10% of this data
- Female => 90% of this data

Division of sales data as per education
Premium:
- Education => Educated (95% of this data)
- Education => Uneducated (5% of this data)
Mid-Range:
- Education => Educated (75% of this data)
- Education => Uneducated (25% of this data)

```python
from random import randint
import pandas as pd
import numpy as np

import sklearn.metrics as metrics
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
def make_empty_dataframe():
    column_names = ["age", "monthly_income", "sex", "education", "product", "quality"]
    df = pd.DataFrame(columns=column_names)

    return df


def premium_mid_range_splitter(total_data_points, split_ratio):
    return int(total_data_points * split_ratio), int(
        total_data_points * (1 - split_ratio)
    )


def make_age_column(no_of_data_points, age_distribution):
    age = []
    for key in age_distribution.keys():
        for _ in range(age_distribution[key] * no_of_data_points // 100):
            age.append(randint(int(key.split("-")[0]), int(key.split("-")[1])))

    return age


def make_monthly_income(no_of_data_points, income_distribution):
    income = []
    for key in income_distribution.keys():
        for _ in range(income_distribution[key] * no_of_data_points // 100):
            income.append(randint(int(key.split("-")[0]), int(key.split("-")[1])))

    return income


def make_education(no_of_data_points, education_distribution):
    education = []
    for _ in range(education_distribution["uneducated"] * no_of_data_points // 100):
        education.append(0)
    for _ in range(education_distribution["educated"] * no_of_data_points // 100):
        education.append(1)

    return education


def make_gender(no_of_data_points, gender_distribution):
    gender = []
    sample_space_array = []
    sample_space_array += [1] * gender_distribution["male"]
    sample_space_array += [0] * gender_distribution["female"]

    for _ in range(no_of_data_points):
        gender.append(sample_space_array[randint(0, 99)])

    return gender
```

```python
# Earrings buying

def make_premium_dataframe(no_of_premium_data_points):
    premium_df = make_empty_dataframe()

    age_distribution = {"18-30": 10, "30-50": 5, "50-80": 85}
    premium_df["age"] = make_age_column(no_of_premium_data_points, age_distribution)

    income_distribution = {"40-60": 5, "60-80": 5, "80-100": 10, "100-200": 80}
    premium_df["monthly_income"] = make_monthly_income(
        no_of_premium_data_points, income_distribution
    )

    education_distribution = {"uneducated": 5, "educated": 95}
    premium_df["education"] = make_education(
        no_of_premium_data_points, education_distribution
    )

    gender_distribution = {"male": 5, "female": 95}
    premium_df["sex"] = make_gender(no_of_premium_data_points, gender_distribution)

    premium_df["product"] = [3] * no_of_premium_data_points
    premium_df["quality"] = [1] * no_of_premium_data_points
    premium_df['purchased'] = [1] * no_of_premium_data_points

    premium_df.to_csv("jewellry_earrings_premium.csv")

    return premium_df


def make_midrange_dataframe(no_of_midrange_data_points):
    premium_df = make_empty_dataframe()

    age_distribution = {"18-30": 15, "30-50": 75, "50-80": 10}
    premium_df["age"] = make_age_column(no_of_midrange_data_points, age_distribution)

    income_distribution = {"10-20": 5, "20-40": 15, "40-60": 80}
    premium_df["monthly_income"] = make_monthly_income(
        no_of_midrange_data_points, income_distribution
    )

    education_distribution = {"uneducated": 25, "educated": 75}
    premium_df["education"] = make_education(
        no_of_midrange_data_points, education_distribution
    )

    gender_distribution = {"male": 10, "female": 90}
    premium_df["sex"] = make_gender(no_of_midrange_data_points, gender_distribution)

    premium_df["product"] = [3] * no_of_midrange_data_points
    premium_df["quality"] = [0] * no_of_midrange_data_points
    premium_df['purchased'] = [1] * no_of_midrange_data_points

    premium_df.to_csv("jewellry_earrings_midrange.csv")

    return premium_df



def main():
    total_data_points = int(input("Enter total number of data points you need : "))
    split_ratio = 0.5
    no_of_premium_data_points, no_of_midrange_data_points = premium_mid_range_splitter(
        total_data_points, split_ratio
    )
    make_premium_dataframe(no_of_premium_data_points)
    make_midrange_dataframe(no_of_midrange_data_points)


main()

```
